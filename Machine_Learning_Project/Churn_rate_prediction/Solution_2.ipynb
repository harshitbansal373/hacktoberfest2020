# Feature Engineering

```python
dataset.info()
```

```python
# Convert Categorical features
X_cat = dataset[['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus']]
```

```python
X_cat.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python
X_cat_2 = pd.DataFrame()
```

```python
for col in X_cat:
    X_cat_2[col] = label.fit_transform(X_cat[col])
```

```python
X_cat_2.head()
```

```python
X_numerical = dataset[['Age','DailyRate','DistanceFromHome','Education','EnvironmentSatisfaction','HourlyRate', 'JobInvolvement','JobLevel','JobSatisfaction','MonthlyIncome','MonthlyRate', 'NumCompaniesWorked', 'OverTime','PercentSalaryHike', 'PerformanceRating', 'RelationshipSatisfaction','StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear','WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole','YearsSinceLastPromotion', 'YearsWithCurrManager']]
```

```python
X_all = pd.concat([X_cat_2 , X_numerical] , axis=1)
X_all.head()
```

```python
# Let's Normalize the data before feeding to neural network
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X = scaler.fit_transform(X_all)
```

```python
y = dataset['Attrition']
y.head()
```

```python
y.value_counts()
```

# As the dataset is imbalanced,let's balance it!

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler

over = SMOTE(sampling_strategy = 0.20)  # Let's increase the records of minority class by 25 percent
under = RandomUnderSampler(sampling_strategy = 0.5) # Decrease majority class upto ratio such that ratio is 40%
```

```python
X_over ,y_over = over.fit_resample(X,y)
X_new ,y_new = under.fit_resample(X_over,y_over)
print(X_over.shape)
print(X.shape)
print(X_new.shape)
```

# Lets Build the models

```python
# 1. Machine Learning Approach
```

```python
#Splitting data into training and testing
from sklearn.model_selection import train_test_split
X_train,X_test , y_train,y_test = train_test_split(X_new,y_new)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import make_pipeline , Pipeline
```

```python
classifier = RandomForestClassifier()
```

```python
pipe = Pipeline([('classifier',classifier)])
params = {'classifier__n_estimators':[50,100 , 10] , 'classifier__max_depth':[1,3,5]}
```

```python
model = GridSearchCV(pipe , param_grid=params , cv=3)
model.fit(X_train,y_train)
```

```python
model.best_params_
```

```python
classifer = RandomForestClassifier(max_depth=5 , n_estimators = 50)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report , confusion_matrix
cm = confusion_matrix(y_test , y_pred)
print(cm)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# Deep Learning Model
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units = 128,activation='relu',input_shape = (X_new.shape[1], )))
model.add(Dense(units = 128 , activation = 'relu'))
model.add(Dense(units = 1 , activation = 'sigmoid'))
model.summary()
```

```python
model.compile(optimizer='Adam',loss = 'binary_crossentropy',metrics = ['accuracy'])
history = model.fit(X_new , y_new , epochs = 40 , batch_size = 32)
```

```python
y_pred_2 = model.predict(X_test)
y_pred_2 = (y_pred_2 > 0.5)
```

```python
plt.plot(history.history['loss'])
```

```python
cm = confusion_matrix(y_test ,y_pred_2)
print(cm)
```
